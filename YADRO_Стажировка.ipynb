##При f=0.5 Гц##

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sine_wave(frequency, sample_rate, duration):
    t = np.arange(0, duration, 1/sample_rate)
    sine_wave = np.sin(2 * np.pi * frequency * t)
    return t, sine_wave

frequency = 0.5
sample_rate = 100
duration = 2

t, sine_wave = generate_sine_wave(frequency, sample_rate, duration)

plt.plot(t, sine_wave)
plt.title('Синусоида с частотой {:.2f} Гц'.format(frequency))
plt.xlabel('Время (с)')
plt.ylabel('Амплитуда')
plt.grid()
plt.show()
```

```python
from scipy.interpolate import interp1d

# Интерполяция
def interpolate_signal(signal, factor):
    original_indices = np.arange(len(signal))
    new_indices = np.linspace(0, len(signal) - 1, len(signal) * factor)
    interpolator = interp1d(original_indices, signal, kind='linear')
    interpolated_signal = interpolator(new_indices)
    return interpolated_signal

interpolated_signal = interpolate_signal(sine_wave, 2)

plt.plot(np.arange(len(sine_wave)) / sample_rate, sine_wave, label='Исходный сигнал')
plt.plot(np.arange(len(interpolated_signal)) / (sample_rate * 2), interpolated_signal, label='Интерполированный сигнал', linestyle='--')
plt.title('Интерполяция сигнала')
plt.xlabel('Время (с)')
plt.ylabel('Амплитуда')
plt.legend()
plt.grid()
plt.show()
```

```python
def decimate_signal(signal, factor):
    return signal[::factor]  # Выбор каждого n-го элемента

# Децимация сигнала
decimated_sine = decimate_signal(sine_wave, factor=2)
t_decim = t[::2]

plt.figure(figsize=(10, 4))
plt.plot(t, sine_wave, 'b-', label='Исходный сигнал')
plt.plot(t_decim, decimated_sine, 'r*', label='Децимированный (/2)')
plt.title('Децимация сигнала (прореживание)')
plt.xlabel('Время (с)')
plt.ylabel('Амплитуда')
plt.grid()
plt.legend()
plt.show()
```

```python
# Применяем децимацию и интерполяцию
decimated_signal = decimate_signal(sine_wave, 2)
interpolated_after_decimate = interpolate_signal(decimated_signal, 2)

# Вычисляем ошибку
error = sine_wave - interpolated_after_decimate[:len(sine_wave)] # Учитываем длину
mse = np.mean(error**2)  # Среднеквадратичная ошибка (MSE)

plt.figure(figsize=(15, 5))
plt.plot(t, sine_wave, label='Исходный сигнал', alpha=0.5)
plt.plot(t, interpolated_after_decimate[:len(t)], label='Сигнал после децимации и интерполяции', linestyle='--')
plt.title('Сравнение сигналов')
plt.xlabel('Время (с)')
plt.ylabel('Амплитуда')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12.64, 5))
plt.plot(t, error, 'k-', label='Ошибка')
plt.title('Ошибка восстановления')
plt.xlabel('Время (с)')
plt.ylabel('Разница')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

print(f'Среднеквадратичная ошибка (MSE) между исходным сигналом и сигналом после обработки: {mse:.6f}')
```

##Все графики сразу при частотах f = [0.5, 5, 25, 40] Гц##

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

def generate_sine_wave(frequency, sample_rate, duration):
    t = np.arange(0, duration, 1/sample_rate)
    sine_wave = np.sin(2 * np.pi * frequency * t)
    return t, sine_wave

def interpolate_signal(signal, factor):
    original_indices = np.arange(len(signal))
    new_indices = np.linspace(0, len(signal) - 1, len(signal) * factor)
    interpolator = interp1d(original_indices, signal, kind='linear')
    interpolated_signal = interpolator(new_indices)
    return interpolated_signal

def decimate_signal(signal, factor):
    return signal[::factor]

frequencies = [0.5, 5, 25, 40]
sample_rate = 100
duration = 2

for frequency in frequencies:
    t, sine_wave = generate_sine_wave(frequency, sample_rate, duration)
    interpolated_signal = interpolate_signal(sine_wave, 2)

    decimated_signal = decimate_signal(sine_wave, 2)
    t_decim = t[::2]

    interpolated_after_decimate = interpolate_signal(decimated_signal, 2)

    error = sine_wave - interpolated_after_decimate[:len(sine_wave)]
    mse = np.mean(error**2)

    plt.figure(figsize=(15, 10))

    # Синусоида
    plt.subplot(3, 1, 1)
    plt.plot(t, sine_wave, label='Исходный сигнал', color='b')
    plt.title(f'Синусоида с частотой {frequency:.2f} Гц')
    plt.xlabel('Время (с)')
    plt.ylabel('Амплитуда')
    plt.grid()
    plt.legend()

    # Децимация
    plt.subplot(3, 1, 2)
    plt.plot(t, sine_wave, 'b-', label='Исходный сигнал')
    plt.plot(t_decim, decimated_signal, 'r*', label='Децимированный (/2)')
    plt.title('Децимация сигнала (прореживание)')
    plt.xlabel('Время (с)')
    plt.ylabel('Амплитуда')
    plt.grid()
    plt.legend()

    # Сравнение сигналов
    plt.subplot(3, 1, 3)
    plt.plot(t, sine_wave, label='Исходный сигнал', alpha=0.5)
    plt.plot(t, interpolated_after_decimate[:len(t)], label='Сигнал после децимации и интерполяции', linestyle='--', color='orange')
    plt.title('Сравнение сигналов после обработки')
    plt.xlabel('Время (с)')
    plt.ylabel('Амплитуда')
    plt.grid()
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Вывод ошибки
    print(f'Среднеквадратичная ошибка (MSE) для частоты {frequency:.2f} Гц: {mse:.6f}')
```

##Результат:
При генерации синусоидального сигнала с частотой 0.5 Гц и частотой дискретизации 100 Гц мы можем наблюдать, что децимация сигнала и последующая интерполяция приводят к потере информации о высоких частотах, что может увеличить ошибку. Также можно заметить, что по мере приблежения к значению частоты дискритизации среднеквадратичная ошибка росла

Децимация: Уменьшает количество выборок, что может привести к потерям в детализации сигнала.

Интерполяция: Восстанавливает выборки, используя имеющиеся данные, но, поскольку исходный сигнал был нечетным после децимации, точные значения в интерполированном сигнале могут не совпадать с исходными.
